# Indeksowanie, wycinanie, iterowanie - Indexing, Slicing and Iterating

### Stwórz liste sześcianów liczb od 0 do 9

```python
a = np.arange(10) ** 3
a
```

### wybierz z listy zaznaczone elementy

[ 0, 1, *8, *27, *64, 125, 216, 343, 512, 729]

    

```python
a[2:5]
```

[ *0, 1, *8, 27, *64, 125, 216, 343, 512, 729]

```python
a[:5:2]
```

### przekształć by otrzymać wynik

[-1000, 1,  -1000, 27, -1000,   125,   216,   343,   512, 729]

```python
a[:5:2] = -1000
a
```

### Odwróć listę

```python
a = a[::-1]
a
```

### Sprawdź typ 

```python
a.dtype
# a.astype('float64')
```

### W pętli for wyprintuj dla każdego elementu pierwiastek sześcienny z elementu


```python
# print(type((-1000) ** (1 / 3)))
for x in a:
#     print(x ** (1 / 3))
    print(np.cbrt(x))
```

### Stwórz macierz 10x10 z kwadratami liczb od 1 do 100

```python
b = np.arange(1, 101).reshape((10, 10)) ** 2
print(b)
```

### Stwórz macierz 5x5 gdzie na krawędziach będą 0 a w środku 1

```python
c = np.zeros((5,5))
c[1:-1, 1:-1] = 1
c
```

### Stwórz tabliczkę mnożenia liczb od 0 do 9 w postaci dwuwymiarowej macierzy

```python
x = np.arange(10)
y = np.arange(10).reshape((10, 1))
x * y
```

### Obrót o wskazaną liczbę stopni 

wykorzystaj ndimage.rotate ze scipy

```python
from scipy import ndimage
y = ndimage.rotate(im1, 45)
y[y < 0] = 0
plt.imshow(y)
```

### Czarno biały filtr

```python
sums = im1.sum(axis=2) - 1
sums = sums / 3
wys, szer = sums.shape
sums = sums.reshape((wys, szer, 1))
result = sums * np.ones(3)
plt.imshow(result)
```
